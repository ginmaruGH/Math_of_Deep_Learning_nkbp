## 2.2 合成関数・逆関数

### 図2-6 逆関数のグラフ

```python
# 関数と逆関数
def f(x):
    return (x**2 + 1)

def g(x):
    return (np.sqrt(x - 1))
```

```python
# データの作成
xx1 = np.linspace(0.0, 4.0, 200)
xx2 = np.linspace(1.0, 4.0, 200)
yy1 = f(xx1)
yy2 = g(xx2)
```

```python
plt.figure(figsize=(6, 6))
plt.xlabel('$x$', fontsize=14)
plt.ylabel('$y$', fontsize=14)
plt.ylim(-2.0, 4.0)
plt.xlim(-2.0, 4.0)
plt.grid(True)
plt.plot(xx1, yy1, linestyle='-', c='blue', label='$y=x^2+1$')
plt.plot(xx2, yy2, linestyle='-.', c='red', label='$y=\sqrt{x-1}$')
plt.plot([-2, 4], [-2, 4], color='gray')
plt.plot([-2, 4], [0, 0], color='gray')
plt.plot([0, 0], [-2, 4], color='gray')
plt.legend(fontsize=14)
plt.show()
```

## 2.3 微分と極限

### 2.3.1 微分の定義

#### 図2-7 関数 $f(x) = x^3 - x$ のグラフを拡大したときの様子

```python
def f(x):
    return (x**3 - x)
```

```python
fig = plt.figure(figsize=(12, 18))

delta = 2.0
plt.subplot(311)
x = np.linspace(0.5-delta, 0.5+delta, 200)
y = f(x)
plt.ylim(-3.0/8.0-delta, -3.0/8.0+delta)
plt.xlim(0.5-delta, 0.5+delta)
plt.plot(x, y,  lw=1, c='k')
plt.scatter([0.5], [-3.0/8.0])
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)
plt.grid(True)
plt.title('delta = %.4f' % delta, fontsize=14)

delta = 0.2
plt.subplot(312)
x = np.linspace(0.5-delta, 0.5+delta, 200)
y = f(x)
plt.ylim(-3.0/8.0-delta, -3.0/8.0+delta)
plt.xlim(0.5-delta, 0.5+delta)
plt.plot(x, y,  lw=1, c='k')
plt.scatter([0.5], [-3.0/8.0])
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)
plt.grid(True)
plt.title('delta = %.4f' % delta, fontsize=14)

delta = 0.01
plt.subplot(313)
x = np.linspace(0.5-delta, 0.5+delta, 200)
y = f(x)
plt.ylim(-3.0/8.0-delta, -3.0/8.0+delta)
plt.xlim(0.5-delta, 0.5+delta)
plt.plot(x, y, lw=1, c='k')
plt.scatter([0.5], [-3.0/8.0])
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)
plt.grid(True)
plt.title('delta = %.4f' % delta, fontsize=14)

plt.show()
```

#### 図2-8 関数のグラフ上の2点を結んだ直線の傾き

```python
delta = 2.0
x = np.linspace(0.5-delta, 0.5+delta, 200)
x1 = 0.6
x2 = 1.0
y = f(x)
fig = plt.figure(figsize=(10, 8))
plt.ylim(-1, 0.5)
plt.xlim(0, 1.5)
plt.plot(x, y, lw=1, c='gray')
plt.scatter([x1, x2], [f(x1), f(x2)], c='red', lw=1)
plt.plot([x1, x2], [f(x1), f(x2)], c='blue', lw=1)
plt.plot([x1, x2, x2], [f(x1), f(x1), f(x2)], c='k', lw=1)
# plt.tick_params(labelbottom=False, labelleft=False,
#                 labelright=False, labeltop=False)
plt.tick_params(color='white')
plt.show()
```
