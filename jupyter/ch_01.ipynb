```python
# ライブラリの準備
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams

plt.style.use('seaborn')
rcParams['font.family'] = 'IPAexgothic'
%precision 3
%matplotlib inline
```

## 1.3 はじめての機械学習

```python
# サンプルデータの準備
sampleData1 = np.array([
    [166, 58.7], [176.0, 75.7], [171.0, 62.1], [173.0, 70.4], [169.0, 60.1]
])
print(sampleData1)
```

```python
# 図1-5: 学習データの2次元散布図
for p in sampleData1:
    plt.scatter(p[0], p[1], c='k', s=50)
plt.grid()
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```

```python
# 図1-7: 原点の移動
plt.figure(figsize=(10, 5))
for p in sampleData1:
    plt.scatter(p[0], p[1], c='k', s=50)
plt.grid()
plt.plot([0, 0], [-10, 80], c='k', lw=1)
plt.plot([171, 171], [-10, 80], c='k')
plt.plot([-10, 180], [0, 0], c='k', lw=1)
plt.plot([-10, 180], [65.4, 65.4], c='k')
plt.xlim(-10, 180)
plt.ylim(-10, 80)
plt.show()
```

```python
# 平均値の計算
means = sampleData1.mean(axis=0)
print(means)
```

```python
# 表1-4: 学習データ3
# 座標系を平均値=0に変換する
sampleData2 = sampleData1 - means
print(sampleData2)
```

```python
# 図1-8: 新しい座標系での散布図
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111)

for p in sampleData2:
    plt.scatter(p[0], p[1], c='k', s=50)
ax.grid()
ax.set_xlabel('$X$')
ax.set_ylabel('$Y$')
ax.plot([-6, 6], [0, 0], c='k')
ax.plot([0, 0], [-7.5, 11], c='k')
ax.set_xlim(-5.2, 5.2)
plt.show()
```

```python
# 予測関数の定義
def L(W0, W1):
    return(5*W0**2 + 58*W1**2 - 211.2*W1 + 214.96)
```

```python
# 図1-9: L(0, W1)のグラフ
plt.figure(figsize=(6, 6))
W1 = np.linspace(0, 4, 501)
#plt.ylim(1,3)
plt.plot(W1, L(0, W1))
plt.scatter(1.82, 22.69, s=30)
plt.xlabel('$W_1$')
plt.ylabel('$L(0,W_1)$')
plt.grid()
plt.xlim(0, 3.5)
plt.ylim(0, 200)
plt.show()
```

```python
def pred1(X):
    return 1.82*X
```

```python
# 図1-10: 散布図と回帰直線(座標変換後)
for p in sampleData2:
    plt.scatter(p[0], p[1], c='k', s=50)
X = np.array([-6, 6])
plt.plot(X, pred1(X), lw=1)
plt.plot([-6, 6], [0, 0], c='k')
plt.plot([0, 0], [-11, 11], c='k')
plt.xlim(-5.2, 5.2)
plt.grid()
plt.xlabel('$X$')
plt.ylabel('$Y$')
plt.show()
```

```python
def pred2(x):
    return 1.82*x - 245.9
```

```python
# 図1-11: もとの座標での散布図と予測式のグラフ
for p in sampleData1:
    plt.scatter(p[0], p[1], c='k', s=50)
x = np.array([166, 176])
plt.plot(x, pred2(x), lw=1)
plt.grid()
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```
